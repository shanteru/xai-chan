```python
all_x_batches = []
all_y_batches = []
for patient_id, magnification, item_dict, binary_label, multi_label in tqdm(test_loader):
    all_x_batches.append(item_dict[magnification[0]])
    all_y_batches.append(binary_label)

x_all = torch.cat(all_x_batches, dim=0)
y_all = torch.cat(all_y_batches, dim=0)

with torch.no_grad():
    a_batch = quantus.explain(model, x_all.cuda(), y_all.cuda(), canonizer=canonizer, composite=composite, attributor=attributor)
```
